```python
nb_train_segmented_image = 0
train_images_seg_exist=[]
for i, image_name in enumerate(df_train.ID): #assuming gif
    filename_seg = os.path.join(train_img_path, image_name+ '_seg' + '.png')
    if exists(filename_seg):
        nb_train_segmented_image+=1
        train_images_seg_exist.append(1)
    else: train_images_seg_exist.append(0)
        
train_images = torch.zeros(df_train.shape[0], 3, 256, 256, dtype=torch.float)
train_images_seg = torch.zeros(nb_train_segmented_image, 1, 256, 256, dtype=torch.float)
transform = transforms.Compose([transforms.Resize((256, 256)), transforms.ToTensor()])

# top, right, bottom, left
border = (10, 10, 10, 10)

k=0
for i, image_name in enumerate(df_train.ID): #assuming gif
    filename = os.path.join(train_img_path, image_name + '.jpg')
    im = Image.open(filename)
    x, y, w, h = clear_black_border(im)
    if x!=0:
        im = im.crop((y,x,h,w))
    im = ImageOps.expand(im, border=border, fill="black")
    train_images[i] = transform(im)
    filename_seg = os.path.join(train_img_path, image_name+ '_seg' + '.png')
    if exists(filename_seg):
        im_seg=Image.open(filename_seg)
        if x!=0:
            im_seg = im_seg.crop((y,x,h,w))
        im_seg = ImageOps.expand(im_seg, border=border, fill="black")
        im_seg = im_seg.resize(image_size)
        train_images_seg[k] = transform(im_seg)
        k+=1
        im_seg.close()
    
train_images = np.asarray(train_images)
```

```python
start = time.time()

Nb_images_train = df_train[df_train.seg_img==1].shape[0]
img_ID = df_train[df_train.seg_img==1].ID
index_in_numpy_arr =  img_ID.index
seg_img_predic_train = []
sim_train = []


print('In the train set')
print('------------------------------------------------------------------------------------------------')

for i,ind in enumerate(index_in_numpy_arr):
    
    indice_seg = np.where(index_in_numpy_arr==ind)
    img = train_images[ind].T
    
    im_Seg_bool = train_images_seg[i][0].T == 1 
    im_Seg_expand = np.expand_dims(im_Seg_bool, axis=2)
    im_mul_mask = (im_Seg_expand*img) # pixel-wise product
    mask= segment(img)
    
    mask = resize(mask, img[:,:,0].shape)
    similarity = 1-dice(mask.flatten(), im_Seg_bool.flatten())
    sim_train.append(similarity)
    seg_img_predic_train.append(mask)
    if i%200==0:
        fig = plt.figure(figsize=(10, 10))
        axs = fig.subplots(nrows=1, ncols=3)
        
        mask = np.dstack((mask, mask, mask))
        masked_image = mask*img
        axs[0].imshow(img)
        axs[1].imshow(masked_image)
        axs[2].imshow(im_Seg_bool, cmap='gray')
        plt.show()
        print(f'{i} images in {time.time() - start} s with an IOU: {np.array(sim_train).mean()}')
train_IOU = np.mean(sim_train)

print(f'Test on the given {Nb_images_train} segmented images in the train set:')
print(f'IOU: {train_IOU}')
```

#### Prepare the data

```python
train_images = (255*train_images).astype(int)
test_images = (255*test_images).astype(int)
```

```python
segmented_images = np.zeros((df_train.seg_img.sum() + df_test.seg_img.sum(), 256, 256, 3), dtype=np.uint8)
masks = np.zeros((df_train.seg_img.sum() + df_test.seg_img.sum(), 256, 256))
index_seg_train = df_train[df_train.seg_img==1].index
index_seg_test = df_test[df_test.seg_img==1].index
```

```python

segmented_images[0:index_seg_train.shape[0]] = train_images.transpose(0,3,2,1)[np.asarray(index_seg_train)]
masks[0:index_seg_train.shape[0]] = np.asarray(train_images_seg).transpose(0,3,2,1)[:,:,:,0]

segmented_images[index_seg_train.shape[0]:] = test_images.transpose(0,3,2,1)[np.asarray(index_seg_test)]
masks[index_seg_train.shape[0]:] = np.asarray(test_images_seg).transpose(0,3,2,1)[:,:,:,0]
```

```python
# Create train, val and test splits
image_seg_train, image_seg_val, mask_train, mask_val = train_test_split(segmented_images, np.uint8(masks), test_size=0.2, random_state=123)
```

```python
input_size = 256
batch_size = 32

seg_train_transform = transforms.Compose([transforms.Resize(input_size),
                                          transforms.GaussianBlur(3, sigma=(.1,.1)),
                                          transforms.ToTensor()
                                          ])

seg_train_transform_target = transforms.Compose([transforms.Resize(input_size),
                                                  transforms.ToTensor()])

seg_val_transform = transforms.Compose([transforms.Resize(input_size),
                                        transforms.ToTensor()
                                        ])

seg_val_transform_target = transforms.Compose([transforms.Resize(input_size),
                                              transforms.ToTensor() ])



kwargs = {'num_workers': 2, 'pin_memory': True} if torch.cuda.is_available() else {}


seg_training_dataset = SegmentationDataSet(inputs=image_seg_train,
                                           targets=mask_train,
                                           transform=[seg_train_transform, seg_train_transform_target])

seg_training_dataloader = data.DataLoader(dataset=seg_training_dataset,
                                          batch_size=batch_size,
                                          shuffle=True, **kwargs)


seg_val_dataset = SegmentationDataSet(inputs=image_seg_val,
                                       targets=mask_val,
                                       transform=[seg_val_transform, seg_val_transform_target])

seg_val_dataloader = data.DataLoader(dataset=seg_val_dataset,
                                     batch_size=batch_size,
                                     shuffle=True, **kwargs)



dataloaders = {'train': seg_training_dataloader, 'val':seg_val_dataloader}
dataset_sizes = {'train': len(image_seg_train),'val': len(image_seg_val)}



```

#### Train the model

```python
import torch
model = torch.hub.load('mateuszbuda/brain-segmentation-pytorch', 'unet',
    in_channels=3, out_channels=1, init_features=32, pretrained=True)

model = model.to(device)


criterion = nn.BCELoss()
# Observe that all parameters are being optimized
optimizer = optim.SGD(model.parameters(), lr=0.005, momentum=0.99)
#optimizer_ft = optim.Adam(model_ft.parameters(), lr=0.0004, betas=[0.5,0.999])
# Decay LR by a factor of 0.1 every 7 epochs
exp_lr_scheduler = lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.3)
```

```python
class MyDatasetRaw(torch.utils.data.Dataset):
    """ISIC 2019 dataset."""
    # classes = {'NV': 0, 'MEL': 1, 'BKL': 2, 'DF': 3, 'SCC': 4, 'BCC': 5, 'VASC': 6, 'AK': 7}
    
    def __init__(self, raw_data, targets, transform=None):
        """
        Args:
        """
        self.raw_data = raw_data
        self.targets = targets
        self.transform = transform        
    
    def __getitem__(self, idx):
        x = Image.fromarray(self.raw_data[idx, :, :, :])
        if self.transform:
            x = self.transform(x)
        y = self.targets.iloc[idx]
        y = np.int64(y)
        id=self.targets.index[idx]
        return {"image": x, "label": y, "Id": id}
            
    def __len__(self):
        return int(len(self.raw_data))
```

```python
from skimage import measure, transform
def shape_asym(mask):
    contours, _  = cv.findContours(mask.astype('int'), cv.RETR_FLOODFILL, cv.CHAIN_APPROX_SIMPLE)
    cnt = contours[0]
    rect = cv.minAreaRect(cnt)
    box = cv.boxPoints(rect)
    box = np.int0(box)
    cX, cY = int(rect[0][0]),int(rect[0][1])
    
    if rect[1][0]>=rect[1][1]:
        img = transform.rotate(mask, rect[2], center=(cX, cY))
    else: 
        img = transform.rotate(mask, 90 + rect[2], center=(cX, cY))
    minor_axe = min(rect[1])
    major_axe = max(rect[1])
    
    
    
    A1_U_A2 = img[int(cY): 0 :-1, :]
    A3_U_A4 = img[int(cY): , :]
    
    l = np.min([A3_U_A4.shape[0], A1_U_A2.shape[0]])
    f1 = np.abs(A1_U_A2[:l, :] - A3_U_A4[:l, :]).sum() / img.sum()
        
 
    A1_U_A4 = img[:, int(cX): 0 :-1]
    A2_U_A3 = img[:, int(cX): ]
    l = np.min([A1_U_A4.shape[1], A2_U_A3.shape[1]])
    f2 = np.abs(A1_U_A4[:, :l] - A2_U_A3[:, :l]).sum() / img.sum()

    return [f1, f2]
```

```python
from skimage import measure, transform
from sklearn.neighbors import KernelDensity
def color_asym(img, mask):
    img = np.uint8(255*255*rescale(img , 0.5, anti_aliasing=True,  order=1, preserve_range=True))
    mask = np.uint8(rescale(mask , 0.5, anti_aliasing=True,  order=1, preserve_range=True))
    
    masked_img = (mask*img)
    cY, cX = mahotas.center_of_mass(masked_img)
    X_plot = np.linspace(0, 255, 256)[:, np.newaxis] 
    S1 = []
    for angle in range(9):
        masked_im = (transform.rotate(masked_img, 10*angle, center=(cX, cY))*255).astype(int)
        
        indices = np.where(masked_im>0)

        A1_U_A2 = masked_im[int(cY) : indices[0].min() : -1, :] 
        A3_U_A4 = masked_im[int(cY) : indices[0].max(), : ]

        kde_A3_U_A4 = KernelDensity(kernel='gaussian', bandwidth=3).fit(A3_U_A4[A3_U_A4>0].reshape((-1,1)))
        C34 = np.exp(kde_A3_U_A4.score_samples(X_plot))

        kde_A1_U_A2 = KernelDensity(kernel='gaussian', bandwidth=3).fit(A1_U_A2[A1_U_A2>0].reshape((-1,1)))
        C12 = np.exp(kde_A1_U_A2.score_samples(X_plot))
        S1.append(np.abs(C12 - C34).sum())
    
    angle = 10*np.argmin(S1)
    masked_im = (transform.rotate(masked_img, angle, center=(cX, cY))*255).astype(int)
    indices = np.where(masked_im>0)
    A1_U_A4 = masked_im[:, int(cX) : indices[1].min() : -1]
    A2_U_A3 = masked_im[:, int(cX) : indices[1].max()]

    
    kde_A1_U_A4 = KernelDensity(kernel='gaussian', bandwidth=3).fit(A1_U_A4[A1_U_A4>0].reshape((-1,1)))
    C14 = np.exp(kde_A1_U_A4.score_samples(X_plot))
    kde_A2_U_A3 = KernelDensity(kernel='gaussian', bandwidth=3).fit(A2_U_A3[A2_U_A3>0].reshape((-1,1)))
    C23 = np.exp(kde_A2_U_A3.score_samples(X_plot))
    S2 = np.abs(C14 - C23).sum()

    
    return [S1[np.argmin(S1)], S2]
```

#### Asymmetry of color-shape (2 features)

```python
def asym_color_shape(img, mask):
    contours, _  = cv.findContours(mask.astype('int'), cv.RETR_FLOODFILL, cv.CHAIN_APPROX_SIMPLE)
    cnt = contours[0]
    rect = cv.minAreaRect(cnt)
    major_axis = max(rect[1])
    m = mask*img*255
    center = mahotas.center_of_mass(m)
    v = []
    for i in range(9):
        th_mask = m*np.logical_and(m<=0.1*i, m>0)
        th_center = mahotas.center_of_mass(th_mask)
        vect_dist = center - th_center
        euc_dist = np.sqrt(vect_dist[0]**2 + vect_dist[1]**2)
        v.append(euc_dist)
    v = np.array(v)/major_axis
    v[np.isnan(v)] = 0
    return [v.mean(), v.std()]
```

### Border, ANOVA-based analysis (3 features)

```python
def Border_Anova(img, mask, window):
    radius = int((window-1)/2)
    img = cv2.copyMakeBorder(img, radius, radius, radius, radius, cv2.BORDER_CONSTANT, value=[0, 0, 0])
    mask = cv2.copyMakeBorder(mask, radius, radius, radius, radius, cv2.BORDER_CONSTANT, value=[0, 0, 0])
    contours, _  = cv.findContours(mask.astype('int'), cv.RETR_FLOODFILL, cv.CHAIN_APPROX_SIMPLE)
    cnt = contours[0]
    wind_mask = np.zeros(mask.shape)
    R=[]
    
    for point in cnt[:,0]:
        wind_mask[point[1] - radius : point[1] + radius, point[0] - radius : point[0] + radius] = 1
        SS_r = (wind_mask*mask*img).sum()
        SS_t = (wind_mask*img).sum()
        R.append(SS_r/SS_t)
        wind_mask[:,:] = 0
    print(contours)
    return np.percentile(R, [25,50,75])


```

```python
from skimage import color
lab = [[x,y,z] for x in range(0, 101, 2) for y in  range(-127, 127, 2) for z in range(-127, 127, 2)]

def color_full_lesion(img, mask, bin=2):
    masked_img = img*np.dstack((mask, mask, mask))
    lab = color.rgb2lab(masked_img)
    L = lab[:,:,0]
    a = lab[:,:,1]
    b = lab[:,:,2]
    
    nL = bin*np.floor(L/bin)
    
    na = bin*np.floor(a/bin)
    
    nb = bin*np.floor(b/bin) 
    
    lab = np.dstack([nL, na, nb]).reshape((-1,3))
    lab_unique, counts = np.unique(lab, axis=0, return_counts=True)
    total_lab_color = 50 * 127 * 127
    
    
    return [counts[1:].mean(), counts[1:].std(),  lab_unique.shape[0]/total_lab_color]

 
```

```python
def color_periph_vs_central(img, mask):
    img = rescale(img , 0.5, anti_aliasing=True,  order=1, preserve_range=True, channel_axis=2)
    mask = np.uint8(rescale(mask , 0.5, anti_aliasing=True,  order=1, preserve_range=True))
    
    m = mask*img.mean(axis=2)
    center = mahotas.center_of_mass(m)
    inner = mask.copy()
    ite = 0
    while True and ite<50:
        ite+=1
        inner = morpho.erosion(inner, footprint=np.array([[1,1,1],[1,1,1],[1,1,1]]))
        if inner.sum()/mask.sum()<=0.7: break
    outer = (1-inner)*mask
    if ite==50: return [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]
    
    
    lab = color.rgb2lab(img)
    
    inner_l = inner*lab[:,:,0]
    inner_a = inner*lab[:,:,1]
    inner_b = inner*lab[:,:,2]
    
    outer_l = outer*lab[:,:,0]
    outer_a = outer*lab[:,:,1]
    outer_b = outer*lab[:,:,2]
    
    f1  = inner_l.mean() - outer_l.mean()
    f2  = inner_a.mean() - outer_a.mean()
    f3  = inner_b.mean() - outer_b.mean()
    
    l_plot = np.linspace(0, 100, 100)[:, np.newaxis] 
    
    kde_inner_l = KernelDensity(kernel='gaussian', bandwidth=1).fit(inner_l[inner_l!=0].reshape((-1,1)))
    kde_outer_l = KernelDensity(kernel='gaussian', bandwidth=1).fit(outer_l[outer_l!=0].reshape((-1,1)))
    dist_inner_l = np.exp(kde_inner_l.score_samples(l_plot))
    dist_outer_l = np.exp(kde_outer_l.score_samples(l_plot))
    l_max = max([dist_outer_l.max(),dist_inner_l.max()])
    overlap_mask = (dist_inner_l / l_max > 1e-2)*(dist_outer_l / l_max > 1e-2)
    overlap = np.min([dist_inner_l*overlap_mask, dist_outer_l*overlap_mask], axis=0)
    f4 = overlap.sum()*1

    
    a_plot = np.linspace(-127, 127, 255)[:, np.newaxis]
    kde_inner_a = KernelDensity(kernel='gaussian', bandwidth=0.5).fit(inner_a[inner_a!=0].reshape((-1,1)))
    kde_outer_a = KernelDensity(kernel='gaussian', bandwidth=0.5).fit(outer_a[outer_a!=0].reshape((-1,1)))
    dist_inner_a = np.exp(kde_inner_a.score_samples(a_plot))
    dist_outer_a = np.exp(kde_outer_a.score_samples(a_plot))
    a_max = max([dist_inner_a.max(), dist_outer_a.max()])
    overlap_mask = (dist_inner_a / a_max > 1e-2)*(dist_outer_a / a_max > 1e-2)
    overlap = np.min([dist_inner_a*overlap_mask, dist_outer_a*overlap_mask], axis=0)
    f5 = overlap.sum()*1

    
    b_plot = np.linspace(-127, 127, 255)[:, np.newaxis]
    kde_inner_b = KernelDensity(kernel='gaussian', bandwidth=3).fit(inner_b[inner_b!=0].reshape((-1,1)))
    kde_outer_b = KernelDensity(kernel='gaussian', bandwidth=3).fit(outer_b[outer_b!=0].reshape((-1,1)))
    dist_inner_b = np.exp(kde_inner_b.score_samples(b_plot))
    dist_outer_b = np.exp(kde_outer_b.score_samples(b_plot))
    b_max = max([dist_inner_b.max(), dist_outer_b.max()])
    overlap_mask = (dist_inner_b / b_max > 1e-2)*(dist_outer_b / b_max > 1e-2)
    overlap = np.min([dist_inner_b*overlap_mask, dist_outer_b*overlap_mask], axis=0)
    f6 = overlap.sum()*1

    return [f1, f2, f3, f4, f5, f6]
```
